```python
#Name: Laura Ye
import pylab as plb

#An array A with 12 elements beginning from number 5, containing consecutive odd numbers
A=plb.arange(12) * 2 + 5

#Compute the simple moving average of an array (SMA) using an adjustable window inside of a function. The function calculating the moving average of the array must take two input arguments:
#-the array itself
#-a window width with a default value =2

#if returnwidth is True, then function will return a tuple with the simple moving average, and the window width. by default this is false.
def sma(arr1, width=2, returnwidth=False):
    
    newarr = plb.array([])
    for i in range(0, len(arr1)-(width-1)):
        newarr = plb.append(newarr,(arr1[i]+arr1[i+width-1])/2)
        
    if returnwidth:
        return (newarr, width)
    else:
        return newarr
    
#Make two function calls
#1. with the array only. save the result to B
B, widthB = sma(A,returnwidth=True)

#2. specifying a window width = 4 save the result to C
C, widthC = sma(A, width=4, returnwidth=True)

#Print the results from the two calls made above using a short description of the data
print('The original array is A:' + str(A))
print('\nMoving average results are:\nB:' + str(B))
print('C:' + str(C))
print('\nWindow widths were: ' + str(widthB) + ' for B and ' + str(widthC) + ' for C.')

#Plot:
#-the sin function of arrays B and C, specifying different: color, linewidth, linestyle, label
plb.figure("Sin of B and C")
fig = plb.gcf()
plb.plot(B, plb.sin(B), color='blue', linewidth=1, linestyle='-', label='sin B')
plb.hold(True)
plb.plot(C, plb.sin(C), color='red', linewidth=2, linestyle='-.', label='sin C')

#-Change the axis using limits
plb.ylim(-1.5, 1.5)

#-Show the grid, legend
plb.grid()
plb.title('Sin of B and C')
plb.legend(['sin B','sin C'], loc='lower left')

plb.show()
plb.close()
```
